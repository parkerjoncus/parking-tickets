#### This notebook modifies the duplicate dataset to add two new features. First, a feature which assigns each duplicate group its own unique ID allowing for easier manipulation of the groups. Second, a feature which identifies each member of a duplicate group as either the first (valid) ticket or repeat ticket.

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
data = pd.read_csv("/Users/ServantezTech/Desktop/cleaned_duplicates.csv", dtype=object)
```

```python
data["group_id"] = "0"
```

```python
groupID = 1
groupDict = dict()

for index, row in data.iterrows():
        
    rawDate = row["issue_date"]
    date = rawDate.split(" ", 2)[0]
    licensePlateNum = row["license_plate_number"]
    code = row["violation_code"]

    key = licensePlateNum + date + code
    
    if key in groupDict:
        row["group_id"] = groupDict[key]
    else:
        row["group_id"] = str(groupID)
        groupDict[key] = str(groupID)
        groupID += 1
```

```python
data = data.sort_values("issue_date")
```

```python
data.head()
```

```python
data["is_repeat"] = "0"
```

```python
repeatDict = dict()

for index, row in data.iterrows():
    
    key = row["group_id"]
    
    if key in repeatDict:
        row["is_repeat"] = 1
    else:
        repeatDict[key] = True
```

```python
data.to_csv("/Users/ServantezTech/Desktop/labeled_duplicates.csv", index=False)
```
